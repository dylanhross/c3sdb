```python
# setup...
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
from matplotlib import pyplot as plt
from matplotlib import rcParams

rcParams['font.size'] = 6

from C3SData.data import C3SD

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# initialize a full dataset
c3sd = C3SD('C3S.db', seed=12345)
c3sd.featurize()

print('X_ shape:', c3sd.X_.shape)
print('y_ shape:', c3sd.y_.shape)
print('features:', c3sd.n_features_)

# compute a 3-component PCA on the complete database using the full feature set
# use the StandardScaler to center and scale the data first
ss = StandardScaler()
X_scaled = ss.fit_transform(c3sd.X_)
pca = PCA(n_components=3)
pc1, pc2, pc3 = pca.fit_transform(X_scaled).T
pc1_evr, pc2_evr, pc3_evr = pca.explained_variance_ratio_
print('explained variance:')
print('    PC1: {:.1f} %'.format(pc1_evr * 100))
print('    PC2: {:.1f} %'.format(pc2_evr * 100))
print('    PC3: {:.1f} %'.format(pc3_evr * 100))
```

```python
kmc = KMeans(n_clusters=4, random_state=12345)
kmc.fit(X_scaled)
```

```python
print([len([l for l in kmc.labels_ if l == n]) for n in range(4)])
```

```python
# plot the PCA projections
fig = plt.figure(figsize=(2, 4))
ax2 = fig.add_subplot(211)
ax4 = fig.add_subplot(212)


# cluster centers
cluster_center_pcs = [pca.transform([_])[0] for _ in kmc.cluster_centers_]
print(cluster_center_pcs)

# color by chemical class label
lab_to_c = [
    '#205f7c',
    '#6a61C5',
    '#ef5675',
    '#ffa600'
]
c = [lab_to_c[_] for _ in kmc.labels_]
lab = [_ for _ in lab_to_c]

# PC1 vs. PC2
ax2.axvline(0, c='k', ls='--', lw=0.5)
ax2.axhline(0, c='k', ls='--', lw=0.5)
ax2.scatter(pc1, pc2, c=c, s=2, alpha=0.2, edgecolors='none')
for d in ['top', 'left', 'bottom', 'right']:
    ax2.spines[d].set_visible(False)
ax2.set_xlabel('PC1 ({:.1f} %)'.format(pc1_evr * 100))
ax2.set_ylabel('PC2 ({:.1f} %)'.format(pc2_evr * 100))

# PC2 vs. PC3
ax4.axvline(0, c='k', ls='--', lw=0.5)
ax4.axhline(0, c='k', ls='--', lw=0.5)
ax4.scatter(pc2, pc3, c=c, s=2, alpha=0.2, edgecolors='none')
for d in ['top', 'left', 'bottom', 'right']:
    ax4.spines[d].set_visible(False)
#ax2.set_title('color by chem_class_label', fontweight='bold')
ax4.set_xlabel('PC2 ({:.1f} %)'.format(pc2_evr * 100))
ax4.set_ylabel('PC3 ({:.1f} %)'.format(pc3_evr * 100))

# add the cluster center projections onto the plots
for i in range(4):
    cpc1, cpc2, cpc3 = cluster_center_pcs[i]
    ax2.scatter(cpc1, cpc2, c=lab_to_c[i], s=12, edgecolors='k', linewidths=0.7)
    ax4.scatter(cpc2, cpc3, c=lab_to_c[i], s=12, edgecolors='k', linewidths=0.7)
    

plt.tight_layout()
plt.savefig('KMC4_PCA_projections.png', dpi=400, bbox_inches='tight')
plt.show()
plt.close()
```

```python
from numpy import mean
fig = plt.figure(figsize=(2, 2))
ax = fig.add_subplot(111)


ccs_by_cluster = [[], [], [], []]
for l, cc in zip(kmc.labels_, c3sd.ccs_):
    ccs_by_cluster[l].append(cc)
mean_cluster_ccs = [mean(_) for _ in ccs_by_cluster]

mz_by_cluster = [[], [], [], []]
for l, m in zip(kmc.labels_, c3sd.mz_):
    mz_by_cluster[l].append(m)
mean_cluster_mz = [mean(_) for _ in mz_by_cluster]


ax.scatter(c3sd.mz_, c3sd.ccs_, c=c, s=2, alpha=0.2, edgecolors='none')

for d in ['top', 'right']:
    ax.spines[d].set_visible(False)
ax.set_xlabel('m/z')
ax.set_ylabel(r'CCS ($^2$)')
ax.set_ylim([100, 380])
ax.set_xlim([50, 1200])

# add the cluster center projections onto the plots
for i in range(4):
    ax.scatter(mean_cluster_mz[i], mean_cluster_ccs[i], 
               c=lab_to_c[i], s=16, edgecolors='k', linewidths=0.75)
    
    

plt.tight_layout()
plt.savefig('KMC4_mz_v_ccs.png', dpi=400, bbox_inches='tight')
plt.show()
```

```python
from numpy import sum, power, sqrt, argmin

# find the central structures in each cluster
d_to_central_cluster = [[], [], [], []]
for i in range(4):
    for sample in X_scaled:
        d_to_central_cluster[i].append(sqrt(sum(power(sample - kmc.cluster_centers_[i], 2))))

print([c3sd.cmpd_[argmin(_)] for _ in d_to_central_cluster])
```
