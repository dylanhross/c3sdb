```python
from sqlite3 import connect
from numpy import sqrt, abs, random
from time import sleep
from sklearn.metrics import mean_squared_error

from lipidccs_pred_scrape import get_session, scrape_token, predict
```

```python
# get the lipids and peptides data from the database
con = connect('C3S.db')
cur = con.cursor()

lipid_smi, lipid_adduct, lipid_ccs = [], [], []
peptide_smi, peptide_adduct, peptide_ccs = [], [], []
qry = 'SELECT smi, adduct, ccs FROM master WHERE smi IS NOT NULL AND chem_class_label="{}"'
for smi, add, ccs in cur.execute(qry.format('lipid')).fetchall():
    lipid_smi.append(smi)
    lipid_adduct.append(add)
    lipid_ccs.append(ccs)
for smi, add, ccs in cur.execute(qry.format('peptide')).fetchall():
    peptide_smi.append(smi)
    peptide_adduct.append(add)
    peptide_ccs.append(ccs)

print([len(lipid_smi), len(lipid_adduct), len(lipid_ccs)])
print([len(peptide_smi), len(peptide_adduct), len(peptide_ccs)])
```

```python
# set up the LipidCCS predictor scraper
sess = get_session()
token = scrape_token(sess, 'prediction')
```

```python
def get_pred_ccs(smis, adducts): 
    pred_ccs = []
    total = len(smis)
    current = 1
    for smi, adduct in zip(smis, adducts):
        print('prediction {} of {} ... '.format(current, total), end='')
        ccs = None
        try:
            mz, ccs = predict(sess, token, smi, adduct)
        except:
            ccs = None
        if ccs:
            print('ok')
        else:
            print('failed')
        pred_ccs.append(ccs)
        current += 1
        sleep(abs(random.normal(2, 1)))
    return pred_ccs
```

```python
# get lipid CCS predictions
lipid_ccs_pred = get_pred_ccs(lipid_smi, lipid_adduct)
```

```python
# get peptide CCS predictions
peptide_ccs_pred = get_pred_ccs(peptide_smi, peptide_adduct)
```

```python
# add in carbohydrates as well just for fun
carb_smi, carb_adduct, carb_ccs = [], [], []
for smi, add, ccs in cur.execute(qry.format('carbohydrate')).fetchall():
    carb_smi.append(smi)
    carb_adduct.append(add)
    carb_ccs.append(ccs)

print([len(carb_smi), len(carb_adduct), len(carb_ccs)])
```

```python
# get carbohydrate CCS predictions
carb_ccs_pred = get_pred_ccs(carb_smi, carb_adduct)
```

```python
print(len([_ for _ in lipid_ccs_pred if _]), 
      len([_ for _ in peptide_ccs_pred if _]), 
      len([_ for _ in carb_ccs_pred if _]))
```

```python
# filter out all of the lipid CCS values for which a LipidCCS prediction was obtained
lipid_ccs_, lipid_ccs_pred_ = [], []
for ccs, pred in zip(lipid_ccs, lipid_ccs_pred):
    if pred:
        lipid_ccs_.append(ccs)
        lipid_ccs_pred_.append(pred)
```

```python
# filter out all of the peptide CCS values for which a peptideCCS prediction was obtained
peptide_ccs_, peptide_ccs_pred_ = [], []
for ccs, pred in zip(peptide_ccs, peptide_ccs_pred):
    if pred:
        peptide_ccs_.append(ccs)
        peptide_ccs_pred_.append(pred)
```

```python
# filter out all of the carb CCS values for which a carbCCS prediction was obtained
carb_ccs_, carb_ccs_pred_ = [], []
for ccs, pred in zip(carb_ccs, carb_ccs_pred):
    if pred:
        carb_ccs_.append(ccs)
        carb_ccs_pred_.append(pred)
```

```python
print(mean_squared_error.__doc__)
```

```python
# RMSE
print('lipid        RMSE: {:.2f}'.format(sqrt(mean_squared_error(lipid_ccs_, lipid_ccs_pred_))))
print('peptide      RMSE: {:.2f}'.format(sqrt(mean_squared_error(peptide_ccs_, peptide_ccs_pred_))))
print('carbohydrate RMSE: {:.2f}'.format(sqrt(mean_squared_error(carb_ccs_, carb_ccs_pred_))))
```
